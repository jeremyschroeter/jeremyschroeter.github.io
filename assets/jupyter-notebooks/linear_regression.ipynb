```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 300

# Set up the parameters
n_samples = 100
covariances = [-10, -5, 0, 5, 10]

# Generate independent random data for x
x = np.linspace(0, 10, n_samples)
y_base = np.random.normal(0, 15, n_samples)

colors = ['r', 'b', 'purple', 'gray', 'orange']
labels = [fr'Cov(x, y) $\approx$ {cov}' for cov in covariances]

plt.figure(figsize=(8, 6))

# Loop through each covariance and plot the data
for cov, color, label in zip(covariances, colors, labels):
    # Generate y values with a covariance structure using a linear combination
    y = y_base + cov * x + np.random.normal(0, 5, n_samples)  # Adding noise to the relationship
    print(np.cov(x, y))
    # Plot the data
    plt.scatter(x, y, c=color, label=label, edgecolor='black')

plt.xlabel('x')
plt.xticks([])
plt.yticks([])
plt.ylabel('y')
# Set plot properties
plt.legend(loc='upper left')
plt.show()
```
